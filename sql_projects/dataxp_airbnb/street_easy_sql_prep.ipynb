```python
import numpy as np
import pandas as pd
import os

from dotenv import load_dotenv
```

```python
# _mar is short for median asking rent
# Note that nyc_three_mar contains data for listings with three or more bedrooms

nyc_studio_mar = pd.read_csv('../../data/dataxp/airbnb_project/raw_data/street_easy/Studio/medianAskingRent_Studio.csv')
nyc_one_mar = pd.read_csv('../../data/dataxp/airbnb_project/raw_data/street_easy/OneBed/medianAskingRent_OneBd.csv')
nyc_two_mar = pd.read_csv('../../data/dataxp/airbnb_project/raw_data/street_easy/TwoBed/medianAskingRent_TwoBd.csv')
nyc_three_mar = pd.read_csv('../../data/dataxp/airbnb_project/raw_data/street_easy/ThreePlusBed/medianAskingRent_ThreePlusBd.csv')
```

### Cleaning the Data

```python
# Zero bedrooms will indicate a studio
# String chosen as data type in order to support '3 plus' bedrooms

nyc_studio_mar['Bedrooms'] = '0'
nyc_one_mar['Bedrooms'] = '1'
nyc_two_mar['Bedrooms'] = '2'
nyc_three_mar['Bedrooms'] = '3 plus'
```

```python
# For future reference. We will need to remove 'All' from the first five entries, but it will be easier to do this
# once all of the DataFrames have been concatenated

nyc_studio_mar['areaName'].tolist()
```

### Moving Bedroom Information to Different Column

```python
cols = nyc_studio_mar.columns.tolist()
```

```python
# This is how we will reorder the columns
# Note that the columns are in the same order in each DataFrame.

cols = cols[:3] + cols[-1:] + cols[3:-1]
```

```python
cols
```

```python
# Checking the reorder.

nyc_studio_mar[cols].head()
```

### Concatenating the Dataframes

```python
# Reorder each column in rental_dfs.

rental_dfs = [
    nyc_studio_mar,
    nyc_one_mar,
    nyc_two_mar,
    nyc_three_mar
]

new_dfs = [ df[cols] for df in rental_dfs ]
```

```python
# Concatenate all reordered DataFrames in new_dfs into a new DataFrame.

sql_df = pd.concat(new_dfs)
```

```python
sql_df
```

### Cleaning the New DataFrame

```python
# Reset the index to match its corresponding row number

sql_df.reset_index(inplace = True)
sql_df.drop('index', axis = 1, inplace = True)
```

```python
sql_df
```

```python
sql_df.rename(columns = {
    'areaName': 'area_name',
    'Borough': 'borough',
    'areaType': 'area_type',
    'Bedrooms': 'bedrooms'
}, inplace = True)
```

```python
# 'All ' replaces the Manhattan area_name rows, and ' All' replaces 'Rockaway All' rows for Queens.

sql_df['area_name'] = sql_df['area_name'].str.replace('All ', '').str.replace(' All', '')
```

```python
sql_df['area_name'].str.contains('All').value_counts()
```

```python
# Alter the data to a format that is friendly for Tableau and SQL.
# list(sql_df.columns[:4] = ['area_name', 'borough', 'area_type', 'bedrooms']

sql_df = sql_df.melt(id_vars = list(sql_df.columns[:4]),
                     var_name = 'date',
                     value_name = 'price')
```

```python
sql_df.head()
```

```python
# Dropping the area_type column

sql_df.drop('area_type', axis = 1, inplace = True)
```

### Cleaning the Date Information

```python
# This is how we will separate the year and month, and give each value its own column.

sql_df['date'].str.split('-')[0]
```

```python
# Put the years and months into their own lists
years = [ date.split('-')[0] for date in sql_df['date'] ]
months = [ date.split('-')[1] for date in sql_df['date'] ]

# Create the columns from the lists
sql_df['year'] = years
sql_df['month'] = months
```

```python
sql_df.info()
```

```python
# If the month begins with 0, then update the value of month to be its second character, i.e. the true month number

for ind in range(len(sql_df)):
    if sql_df.at[ind, 'month'][0] == '0':
        sql_df.at[ind, 'month'] = sql_df.at[ind, 'month'][1]
        
# Give the year and month column integer types
sql_df['year'] = sql_df['year'].astype('int')
sql_df['month'] = sql_df['month'].astype('int')
```

```python
sql_df.info()
```

```python
# Checking the above alteration

sql_df.sample(15, random_state = 4)
```

```python
# Remove the date column

sql_df.drop('date', axis = 1, inplace = True)
```

### Creating the Postgres Table

`sqlalchemy` was chosen as a tool so that a `CREATE TABLE` query would not have to be created manually. It is worth noting that the default column that is created for the numerical columns for this data set is a double precision, which may use more bytes than would be necessary to store the data in Postgres.

```python
load_dotenv()

from sqlalchemy import create_engine
engine = create_engine(os.getenv('ENGINE'))
sql_df.to_sql('street_easy_nyc', engine, index = False)
```
